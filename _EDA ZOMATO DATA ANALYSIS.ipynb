```python
#group by on the basis of aggregragte rattings ,rating colours, rating text #

df1.groupby(["Aggregate rating","Rating color","Rating text"]).size()
```

```python
#rename the above column as rating count previously there was no name for thjis columns 
df1.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns={0:"Rating counts"})
```

```python
Rating = df1.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns={0:"Rating counts"})
```

```python
# Group by 'Aggregate rating' and count the occurrences
rating_counts = df1.groupby('Aggregate rating').size().reset_index(name='Counts')

# Prepare the data for the pie chart
Ratings = rating_counts['Counts']
Aggregate_rating = rating_counts['Aggregate rating']

# Plotting the pie chart
plt.figure(figsize=(8,8))
plt.pie(Ratings, labels=Aggregate_rating, autopct="%.f%%")
plt.title("Distribution of Aggregate Ratings")
plt.show()
```
